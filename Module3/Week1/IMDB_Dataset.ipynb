### **Apply Function**

```python
def rating_group(rating):
    if rating >= 8.0:
        return 'Good'
    elif rating >= 0.5:
        return 'Average'
    else:
        return 'Bad'
    
data['Classify'] = data['Rating'].apply(rating_group)
```
